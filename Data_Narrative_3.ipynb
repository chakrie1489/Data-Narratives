#Question 1

What is the joint probability distribution of the number of double faults and the outcome of a match?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('AusOpen-men-2013.csv')
df = df.replace('NA', '')
df.to_csv('AusOpen-men-2013-filtered.csv', index = False)


joint_counts = pd.crosstab(df['DBF.1'], df['Result'])
total_matches = len(df)
joint_probabilities = joint_counts / total_matches
print("Joint Probabilities:", joint_probabilities)

sns.heatmap(joint_probabilities, annot=True)
plt.xlabel('Match Result')
plt.ylabel('Number of Double Faults')
plt.title('Joint Probability Distribution of Number of Double Faults and Match Result')
plt.show()
```

#Question 2

Suppose we have two random variables representing the first serve percentage (FSP) of Player 1 and Player 2 in the AusOpen-women-2013.csv dataset. Plot the joint probability density function of these two random variables and calculate the covariance between them.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('AusOpen-women-2013.csv')
df = df.replace('NA', '')
df.to_csv('AusOpen-women-2013-filtered.csv', index = False)

fsp1 = df['FSP.1']
fsp2 = df['FSP.2']
sns.jointplot(x=fsp1, y=fsp2, kind='kde')
plt.show()
covariance = np.cov(fsp1, fsp2)[0, 1]
print(f'The covariance between FSP.1 and FSP.2 is {covariance:.2f}')
```

#Question 3

What is the correlation coefficient between the first serve percentage (FSP) and the number of aces (ACE) for both players? Plot a scatter plot with a regression line to visualize the relationship.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('FrenchOpen-men-2013.csv')
df = df.replace('NA', '')
df.to_csv('FrenchOpen-men-2013-filtered.csv', index = False)
corr1 = df['FSP.1'].corr(df['ACE.1'])
corr2 = df['FSP.2'].corr(df['ACE.2'])
print(f'Correlation between FSP and ACE for Player 1: {corr1:.2f}')
print(f'Correlation between FSP and ACE for Player 2: {corr2:.2f}')

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.regplot(x='FSP.1', y='ACE.1', data=df, ax=ax[0])
sns.regplot(x='FSP.2', y='ACE.2', data=df, ax=ax[1])
ax[0].set_title('Player 1')
ax[1].set_title('Player 2')
plt.show()
```

#Question 4

What is the Pearson correlation coefficient between the percentage of first serves made by a player and the percentage of first serve points won by that player in the 2013 French Open Women's tournament?

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

data = pd.read_csv('FrenchOpen-women-2013.csv')
df = df.replace('NA', '')
df.to_csv('FrenchOpen-women-2013-filtered.csv', index = False)
fsm1 = data['FSP.1']
fsw1 = data['FSW.1']

plt.hexbin(fsm1, fsw1, gridsize=20, cmap='Blues')
plt.xlabel('Percentage of First Serves Made by Player 1')
plt.ylabel('Percentage of First Serve Points Won by Player 1')
plt.title('Correlation between First Serve Stats')
plt.colorbar()
plt.show()

corr, _ = pearsonr(fsm1, fsw1)
print('Pearson correlation coefficient:', corr)
```

#question 5

What is the covariance between the first serve percentage (FSP) and the total points won (TPW) for both players in the US Open Mens tournament in 2013?

```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('USOpen-men-2013.csv')
df = df.replace('NA', '')
df.to_csv('USOpen-men-2013-filtered.csv', index = False)
cov1 = df[['FSP.1', 'TPW.1']].cov().iloc[0, 1]
cov2 = df[['FSP.2', 'TPW.2']].cov().iloc[0, 1]
print(f'Covariance between FSP and TPW for Player 1: {cov1:.2f}')
print(f'Covariance between FSP and TPW for Player 2: {cov2:.2f}')

sns.kdeplot(data=df, x='FSP.1', y='TPW.1', label='Player 1')
sns.kdeplot(data=df, x='FSP.2', y='TPW.2', label='Player 2')
plt.xlabel('First Serve Percentage')
plt.ylabel('Total Points Won')
plt.legend()
plt.show()
```

#Question 6

What is the probability that a player who has a higher number of aces (ACE) will go on to win the match in each round?

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('USOpen-women-2013.csv')
rounds = df['ROUND'].unique()
probabilities = []
for round in rounds:
    round_df = df[df['ROUND'] == round]
    higher_ace = round_df[round_df['ACE.1'] > round_df['ACE.2']]
    match_winners = higher_ace[higher_ace['Result'] == 1]
    probability = len(match_winners) / len(higher_ace)
    probabilities.append(probability)
    print(f'Round {round}: Probability of winning match given higher number of aces = {probability}')
plt.bar(rounds, probabilities)
plt.xlabel('Round')
plt.ylabel('Probability of winning match given higher number of aces')
plt.show()
```

#Question 7

What is the covariance between the number of aces hit by a player and the number of games they win in a match?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("Wimbledon-men-2013.csv")
df["games_won_1"] = df["ST1.1"] + df["ST2.1"] + df["ST3.1"] + df["ST4.1"] + df["ST5.1"]
df["games_won_2"] = df["ST1.2"] + df["ST2.2"] + df["ST3.2"] + df["ST4.2"] + df["ST5.2"]
df["total_aces_1"] = df["ACE.1"]
df["total_aces_2"] = df["ACE.2"]
sns.regplot(x="total_aces_1", y="games_won_1", data=df, label="Player1")
sns.regplot(x="total_aces_2", y="games_won_2", data=df, label="Player2")
plt.xlabel("Total Aces")
plt.ylabel("Number of Games Won")
plt.legend()
plt.show()
covariance = df[["total_aces_1", "games_won_1"]].cov().iloc[0, 1]
print("Covariance between total aces and number of games won:", covariance)
```

#Question 8

What is the probability that a player who has a higher ratio of winners to unforced errors (WNR/UFE) will win the match?

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('Wimbledon-women-2013.csv')
df['WNR_UFE_Ratio1'] = df['WNR.1'] / df['UFE.1']
df['WNR_UFE_Ratio2'] = df['WNR.2'] / df['UFE.2']
df['HigherWNR_UFE_RatioWon'] = (df['WNR_UFE_Ratio1'] > df['WNR_UFE_Ratio2']) == (df['Result'] == 1)
prob = df['HigherWNR_UFE_RatioWon'].mean()
print(f'Probability that a player with a higher ratio of winners to unforced errors will win the match: {prob:.2f}')
x = df['WNR_UFE_Ratio1'] - df['WNR_UFE_Ratio2']
y = df['Result']
sns.kdeplot(x=x, y=y)
plt.xlabel('Difference in Ratio of Winners to Unforced Errors (Player 1 - Player 2)')
plt.ylabel('Match Result (1 if Player 1 won, 0 if Player 2 won)')
plt.show()
```
