Most number of books are written in which language?

```python
import pandas as pd
import matplotlib.pyplot as plt

books = pd.read_csv('books.csv')

language = books['language_code']

a = languages.value_counts()
#a is number of books in each language

most_popular_language = a.index[0]
print('The language in which most number of books written is:', most_popular_language)

a.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Distribution of Books by Language')
plt.show()
```

Most of the top 100 books were written in which year?

```python
import pandas as pd
import matplotlib.pyplot as plt
books = pd.read_csv('books.csv')

top_100_books= books.sort_values('average_rating', ascending=False).head(100)

year= top_100_books['original_publication_year']
x = year.value_counts() #x is number of books published in each year

x.plot(kind='bar')
plt.xlabel('Published Year')
plt.ylabel('No. of Books')
plt.title('No.of Top 100 Books by Publication Year')
plt.show()
```

what are trends in the average ratings of jk rowling books?

```python
import pandas as pd
import matplotlib.pyplot as plt
books = pd.read_csv('books.csv')
a = books.loc[books['authors'] == 'J.K. Rowling'] #a represents the books written by jk rowling
jk_rowling_books = a.sort_values('original_publication_year')

plt.plot(a['original_publication_year'], jk_rowling_books['average_rating'])
plt.xlabel('Publication Year')
plt.ylabel('Average Rating')
plt.title('Average Ratings of J.K. Rowling Books Over Time')
plt.show()
```

what number of books that are written by J.K.Rowling are rated above 4.0?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
books = pd.read_csv('books.csv')
a = books[books['authors'] == 'J.K. Rowling'] #a shows the books written by jk rowling
b = np.sum(((a['average_rating']) > '4.0')) #b is number of jk rowling bookks with average rating greater than 4
print(f"{b} books written by J.K. Rowling are rated above 4.0.")
a['average_rating'].hist()
plt.xlabel('Average Rating')
plt.ylabel('Number of Books')
plt.yticks([1,2,3])
plt.title('Distribution of Ratings for J.K. Rowling Books')
plt.show()
```

Which book was marked to read the most number of times?

```python
import pandas as pd
import matplotlib.pyplot as plt
to_read = pd.read_csv('to_read.csv')
book_counts = to_read.groupby('book_id').count()
sorted_counts = book_counts.sort_values(by='user_id', ascending=False)
l = sorted_counts.index[0] #l is the book_id of book with most to read markings
a = pd.read_csv('books.csv')
m = a.loc[a['book_id'] == l, 'title'].values[0] #m is the title of the book with most to read markings
print(f"The book marked 'to read' the most number of times is '{m}' with {sorted_counts.iloc[0, 0]} occurrences.")
book_counts.plot(kind='hist', y='user_id', bins=50)
plt.xlabel('Number of "To Read" occurrences')
plt.ylabel('Number of Books')
plt.title('Distribution of "To Read" occurrences')
plt.show()
```
